# Qus 1

```python
# Line Plot:
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y)
```

```python
# Scatter plot
plt.scatter(x,y,color="red",marker="o")
```

```python
# Histogram
import numpy as np
data=np.random.randn(1000)
plt.hist(data,bins=30,alpha=0.5,color="blue")
```

```python
# Bar Plot
c=["A","B","C","D"]
v=[20,35,30,25]
plt.bar(c,v,color="green")
```

```python
#Pie chart
labels=["A","B","C","D"]
sizes=[20,35,30,25]
plt.pie(sizes,labels=labels,autopct="%1.1f%%",startangle=90)
```
